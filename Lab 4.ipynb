```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Load the data
data = pd.read_csv('Raisin_Dataset_KMEANS.csv')

# Step 2: Select the features(X) for clustering, excluding target Y('Class') and 'id'
X = data.drop(columns=['Class', 'id'])

# Step 3: (Skipped) Standardize the features
# If data is already standardized, proceed to the next step

# Step 4: Run K-Means for a range of cluster numbers and calculate SSE
SSE = []
k_range = range(1, 15)  # you can adjust the range as needed
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    SSE.append(kmeans.inertia_)  # Inertia is the SSE

# Step 5: Plot the SSE to create the elbow chart
plt.figure(figsize=(10, 6))
plt.plot(k_range, SSE, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```
