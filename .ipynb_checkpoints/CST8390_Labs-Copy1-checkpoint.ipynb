```python
# Calculate summary statistics
summary_stats = employees['Salary'].describe()
print(summary_stats)

# Visualize the distribution using a boxplot
employees['Salary'].plot(kind='box')

# Calculate Z-score
z_scores = (employees['Salary'] - employees['Salary'].mean()) / employees['Salary'].std()

# Define outliers based on Z-score threshold
outliers_z = employees[abs(z_scores) > 3]

# Calculate IQR
Q1 = employees['Salary'].quantile(0.25)
Q3 = employees['Salary'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers based on IQR method
outliers_iqr = employees[(employees['Salary'] < Q1 - 1.5 * IQR) | (employees['Salary'] > Q3 + 1.5 * IQR)]

# Print or visualize the outliers
print("Outliers based on Z-score:")
print(outliers_z)
print("Outliers based on IQR method:")
outliers_iqr
```

```python
X =  wine.drop('class', axis=1)
y = wine['class']

# 打印前几行数据
print("first few lines of label(y)：")
print(y.head())

print("\nfirst few lines of features(X)：")
X.head()
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
wine.shape
```

```python
# 定义KFold参数
kfold = 10
kf = KFold(n_splits=kfold, shuffle=True, random_state=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KNN模型
k_values = [1, 3, 5, 9, 11]  # 选择你需要的k值
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # 使用交叉验证进行预测
    y_pred = cross_val_predict(knn, X_scaled, y, cv=kf)
    
    # 生成混淆矩阵
    conf_matrix = confusion_matrix(y, y_pred)
    
    # 打印分类报告
    class_report = classification_report(y, y_pred)
    print(f"\n=== Classification Report (k={k}) ===")
    print(class_report)
    
    # 计算和打印准确率
    accuracy = accuracy_score(y, y_pred)
    print(f'\n准确率: {accuracy}')

    # 打印混淆矩阵
    print(f"=== Confusion Matrix  (k={k}) ===")
    print(conf_matrix)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# KNN模型
k_values = [1, 3, 5, 9, 11]  # 选择你需要的k值
for k in k_values:
    # 定义KNN模型
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # 训练模型
    knn.fit(X_train_scaled, y_train)
    
    # 使用测试集进行预测
    y_pred = knn.predict(X_test_scaled)
    
    # 生成混淆矩阵
    conf_matrix = confusion_matrix(y_test, y_pred)
    
    # 打印混淆矩阵
    print(f"=== Confusion Matrix (k = {k})===")
    print(conf_matrix)
    
    # 打印分类报告
    class_report = classification_report(y_test, y_pred)
    print(f"\n=== Classification Report (k = {k})===")
    print(class_report)
    
    # 计算和打印准确率
    accuracy = accuracy_score(y_test, y_pred)
    print(f'\n准确率: {accuracy}')
```

```python
# Function to detect outliers using IQR
def detect_outliers(df, numeric_columns):
    outliers = pd.DataFrame()
    extremers = pd.DataFrame()
    
    for column in numeric_columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        OF = 3  # Outlier Factor
        EVF = 6  # Extreme Value Factor
        
        outlier_lower_bound = Q1 - OF * IQR
        outlier_upper_bound = Q3 + OF * IQR

        extremer_lower_bound = Q1 - EVF * IQR
        extremer_upper_bound = Q3 + EVF * IQR

        print(column + " extremer range:", extremer_lower_bound, " - ", extremer_upper_bound)
        
        column_outliers = df[((df[column] <= extremer_upper_bound) & (df[column] > outlier_upper_bound)) | ((df[column] < outlier_lower_bound) & (df[column] >= extremer_lower_bound))]
        outliers = pd.concat([outliers, column_outliers])

        column_extremers = df[(df[column] < extremer_lower_bound) | (df[column] > extremer_upper_bound)]
        extremers = pd.concat([extremers, column_extremers])
    
    return extremers.drop_duplicates(), outliers.drop_duplicates()
```

```python
extremers, outliers = detect_outliers(df_unique, numeric_columns)
extremers
```
